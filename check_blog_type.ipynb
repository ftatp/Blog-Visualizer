```python
# 원본
def check_blog_type(test_text):
    detect_dbdbdeep_item = ['소정의원고료를지원받아작성하였습니다.', '의료광고포스팅입니다', '지원받아작성하였습니다', '지원받아작성되었습니다', '마케팅이즈', '원고료']
    experience_item = ['제품을제공받아', '당첨이되어서', '당첨된제품', '무상지원받아', '르뷰', '직접검증']
    ls_dbdbdeep = 0
    ls_experience = 0
    
    # dbdbdeep 찾기
    for item in detect_dbdbdeep_item:
        p = re.compile(item)
        if p.search(test_text):
            print ("dbdbdeep!!") # 제거 가능, 확인용
            print ("item: ", item, end = " ")  # 제거 가능, 확인용
            ls_dbdbdeep = 1
            break
    
    # 체험형 찾기
    if ls_dbdbdeep == 0:
        for item in experience_item:
            p = re.compile(item)
            if p.search(test_text):
                print ("experience") # 제거 가능, 확인용
                print ("item: ", item, end = " ")  # 제거 가능, 확인용
                ls_experience = 1
                break
    
    if ls_dbdbdeep == 1:
        return 0
    
    elif ls_experience == 1:
        return 1

    else:
        print ("else")
        return 2
```
